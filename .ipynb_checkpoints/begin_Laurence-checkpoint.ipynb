```python
import pandas as pd
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("paperboy_instance.xlsx")
df.head()
```

```python
plt.scatter(df["x"], df["y"])
```

```python
df_rb = df[(df["y"] > 300) & (df["x"] > 250)]
df_ro = df[(df["y"] < 300) & (df["x"] > 250)]
df_lb = df[(df["y"] > 300) & (df["x"] < 250)]
df_lo = df[(df["y"] < 300) & (df["x"] < 250)]

print(len(df_rb), len (df_ro), len(df_lb), len(df_lo))

```

```python
df_rb = df_rb.reset_index(drop=True)
df_rb
```

```python
dist = {}
j = 0
k = 1
n = len(df_rb)
while j < n-1:
    d = abs(df.loc[j,"x"] - df.loc[k,"x"]) + abs(df.loc[j,"y"] - df.loc[k,"y"])
    #d = m.sqrt((df.loc[j, "x"] - df.loc[k, "x"])**2 + (df.loc[j, "y"] - df.loc[k, "y"])**2)
    dist[(j, k)] = d
    dist[(k,j)] = d
    k = k + 1
    if k == n:
        j = j + 1
        #print (j)
        k = j + 1
    
#print (dist)
```

```python
#start op punt 1, laat python uit mn dict de kortste distance pakken waar het cijfer instaat, mmaar hij mag m niet dubbel gebruiken, 
#totaal variabele om totale afstand te berekenen
totaal = 0
kortste = 1000 
naar_stad = 0 
i = 0
j = [0]
n = len(df_rb)
while i < n:
    for k in range(n):
        if j[i-1] !=k:
            if j[i] != k:
                afstand = dist[(k,j[i])]
                #print (afstand, j[i], k)
                if afstand < kortste:
                    kortste = afstand
                    naar_stad = k
    j.append(naar_stad)
    print(f'vanaf punt {j[i-1]} is punt {j[i]} het dichtste bij')
    totaal = totaal + kortste
    print (f'de totale afstand is {totaal}')
    i = i + 1
    kortste = 1000
    k = 0
print (f'de route is {j}')
##code klopt nog niet helemaal, hij gaat vaker terug naar 17 omdat die het dichtste bij is dus moeten iets verzinnen dat hij helemaal niet meer gaat naar
##de punten die in lijst j zijn toegevoegd
```
