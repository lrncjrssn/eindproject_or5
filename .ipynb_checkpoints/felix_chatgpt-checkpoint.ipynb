```python
# ==============================================
# OR5 - Newspaper Delivery
# RQ3: Effect van startoplossing (random vs greedy)
# ==============================================

import pandas as pd
import numpy as np
import random
import math
import time
import matplotlib.pyplot as plt

# -------------------------------------------------------
# Functies voor afstand en evaluatie
# -------------------------------------------------------

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def route_lengte(route, depot):
    """Bereken de totale afstand van één route."""
    if len(route) == 0:
        return 0
    d = 0
    cur = depot
    for klant in route:
        d += manhattan(cur, klant)
        cur = klant
    return d

def makespan(oplossing, depot):
    """Het doel: tijdstip van laatste bezorging (max route-lengte)."""
    return max(route_lengte(r, depot) for r in oplossing)

# -------------------------------------------------------
# Constructieve heuristieken
# -------------------------------------------------------

def greedy_construct(klanten, depot, K=4):
    """
    Greedy constructieve heuristiek:
    voeg telkens de klant toe die het makespan het minst verhoogt.
    """
    routes = [[] for _ in range(K)]
    klanten_sorted = sorted(klanten, key=lambda c: manhattan(c, depot), reverse=True)
    for klant in klanten_sorted:
        beste_r = None
        beste_obj = None
        for i in range(K):
            nieuwe_routes = [list(r) for r in routes]
            nieuwe_routes[i].append(klant)
            obj = makespan(nieuwe_routes, depot)
            if beste_obj is None or obj < beste_obj:
                beste_obj = obj
                beste_r = i
        routes[beste_r].append(klant)
    return routes

def random_construct(klanten, K=4):
    """Random startoplossing: verdeel klanten willekeurig over de routes."""
    klanten_copy = klanten[:]
    random.shuffle(klanten_copy)
    routes = [[] for _ in range(K)]
    for i, klant in enumerate(klanten_copy):
        routes[i % K].append(klant)
    return routes

# -------------------------------------------------------
# Local search: hill climbing met eenvoudige buurten
# -------------------------------------------------------

def local_search(oplossing, depot, iteraties=2000):
    """
    Verbeter de gegeven oplossing via lokale zoekoperaties:
    - intra-route 2-opt
    - inter-route relocate
    - inter-route swap
    """
    sol = [r[:] for r in oplossing]
    beste_obj = makespan(sol, depot)
    K = len(sol)
    
    for it in range(iteraties):
        verbeterd = False

        # --- intra-route 2-opt ---
        r_idx = random.randrange(K)
        route = sol[r_idx]
        n = len(route)
        if n >= 4:
            i = random.randrange(0, n-2)
            j = random.randrange(i+2, n)
            nieuwe_route = route[:i+1] + list(reversed(route[i+1:j+1])) + route[j+1:]
            nieuwe_sol = [r[:] if idx!=r_idx else nieuwe_route for idx,r in enumerate(sol)]
            obj = makespan(nieuwe_sol, depot)
            if obj < beste_obj:
                sol = nieuwe_sol
                beste_obj = obj
                verbeterd = True

        # --- relocate (verplaats klant tussen routes) ---
        a = random.randrange(K)
        b = random.randrange(K)
        if a != b and len(sol[a]) > 0:
            i = random.randrange(len(sol[a]))
            klant = sol[a][i]
            nieuwe_a = sol[a][:i] + sol[a][i+1:]
            pos = random.randrange(len(sol[b]) + 1)
            nieuwe_b = sol[b][:pos] + [klant] + sol[b][pos:]
            nieuwe_sol = [sol[idx] if idx not in (a,b) else (nieuwe_a if idx==a else nieuwe_b) for idx in range(K)]
            obj = makespan(nieuwe_sol, depot)
            if obj < beste_obj:
                sol = nieuwe_sol
                beste_obj = obj
                verbeterd = True

        # --- swap (verwissel klanten tussen routes) ---
        a, b = random.sample(range(K), 2)
        if len(sol[a]) > 0 and len(sol[b]) > 0:
            i = random.randrange(len(sol[a]))
            j = random.randrange(len(sol[b]))
            na = sol[a][:]; nb = sol[b][:]
            na[i], nb[j] = nb[j], na[i]
            nieuwe_sol = [sol[idx] if idx not in (a,b) else (na if idx==a else nb) for idx in range(K)]
            obj = makespan(nieuwe_sol, depot)
            if obj < beste_obj:
                sol = nieuwe_sol
                beste_obj = obj
                verbeterd = True

    return sol, beste_obj

# -------------------------------------------------------
# Experimentele setup voor RQ3
# -------------------------------------------------------

def run_experiment(aantal_inst=3, iteraties=4000, klanten_per_inst=120, K=4):
    resultaten = []
    for seed in [101, 202, 303][:aantal_inst]:
        random.seed(seed)
        np.random.seed(seed)
        depot = (250, 250)
        klanten = [(int(x), int(y)) for x, y in np.random.randint(0, 501, size=(klanten_per_inst, 2))]

        # --- GREEDY start ---
        greedy_start = greedy_construct(klanten, depot, K)
        greedy_start_obj = makespan(greedy_start, depot)
        t0 = time.time()
        greedy_final, greedy_final_obj = local_search(greedy_start, depot, iteraties)
        t_greedy = round(time.time() - t0, 3)

        # --- RANDOM start (beste van 3) ---
        random_runs = []
        best_rand_obj = None
        for r in range(3):
            rs = random_construct(klanten, K)
            start_obj = makespan(rs, depot)
            sol_r, obj_r = local_search(rs, depot, iteraties)
            random_runs.append((start_obj, obj_r))
            if best_rand_obj is None or obj_r < best_rand_obj:
                best_rand_obj = obj_r
        t_random = round(time.time() - t0, 3)

        resultaten.append({
            "Instance": seed,
            "Greedy start": greedy_start_obj,
            "Greedy eind": greedy_final_obj,
            "Random beste start": min(r[0] for r in random_runs),
            "Random beste eind": best_rand_obj,
            "Tijd Greedy (s)": t_greedy,
            "Tijd Random (s)": t_random,
            "Verbetering Greedy": greedy_start_obj - greedy_final_obj,
            "Verbetering Random": min(r[0] for r in random_runs) - best_rand_obj
        })

    return pd.DataFrame(resultaten)

# -------------------------------------------------------
# Run de experimenten voor RQ3
# -------------------------------------------------------

df_res = run_experiment()
display(df_res)

# Optioneel: grafische vergelijking
plt.figure(figsize=(6,4))
plt.bar(df_res["Instance"]-0.15, df_res["Greedy eind"], width=0.3, label="Greedy start → eind")
plt.bar(df_res["Instance"]+0.15, df_res["Random beste eind"], width=0.3, label="Random start → eind")
plt.xlabel("Instantie (seed)")
plt.ylabel("Makespan (tijdseenheden)")
plt.legend()
plt.title("Vergelijking Random vs Greedy start (RQ3)")
plt.show()
```

```python
# ==============================
# OR5 - Newspaper Delivery
# RQ3: Effect van startoplossing
# ==============================

import pandas as pd
import numpy as np
import random
import math
import time
import matplotlib.pyplot as plt

# -------------------------------
# Hulpfuncties
# -------------------------------

def manhattan(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def route_lengte(route, depot):
    dist = 0
    cur = depot
    for klant in route:
        dist += manhattan(cur, klant)
        cur = klant
    return dist

def makespan(routes, depot):
    return max(route_lengte(r, depot) for r in routes)

# -------------------------------
# Constructieve heuristieken
# -------------------------------

def greedy_start(klanten, depot, k=4):
    routes = [[] for _ in range(k)]
    klanten_sorted = sorted(klanten, key=lambda c: manhattan(c, depot), reverse=True)
    for klant in klanten_sorted:
        beste_r = None
        beste_obj = 10**9
        for i in range(k):
            temp = [list(r) for r in routes]
            temp[i].append(klant)
            obj = makespan(temp, depot)
            if obj < beste_obj:
                beste_obj = obj
                beste_r = i
        routes[beste_r].append(klant)
    return routes

def random_start(klanten, k=4):
    random.shuffle(klanten)
    routes = [[] for _ in range(k)]
    for i, klant in enumerate(klanten):
        routes[i % k].append(klant)
    return routes

# -------------------------------
# Eenvoudige local search
# -------------------------------

def local_search(routes, depot, iters=1000):
    best = [list(r) for r in routes]
    best_obj = makespan(best, depot)
    k = len(best)

    for _ in range(iters):
        a = random.randrange(k)
        b = random.randrange(k)
        if a != b and len(best[a]) > 0:
            i = random.randrange(len(best[a]))
            klant = best[a][i]
            new_a = best[a][:i] + best[a][i+1:]
            pos = random.randrange(len(best[b])+1)
            new_b = best[b][:pos] + [klant] + best[b][pos:]
            new_routes = [best[x] if x not in (a,b) else (new_a if x==a else new_b) for x in range(k)]
            obj = makespan(new_routes, depot)
            if obj < best_obj:
                best = new_routes
                best_obj = obj
    return best, best_obj

# -------------------------------
# Experiment RQ3
# -------------------------------

resultaten = []
zaden = [101, 202, 303]

for seed in zaden:
    random.seed(seed)
    np.random.seed(seed)

    depot = (250, 250)
    klanten = [(int(x), int(y)) for x, y in np.random.randint(0, 501, (120, 2))]

    # Greedy
    start_g = greedy_start(klanten[:], depot)
    start_g_obj = makespan(start_g, depot)
    t0 = time.time()
    eind_g, eind_g_obj = local_search(start_g, depot, iters=2000)
    tijd_g = round(time.time() - t0, 2)

    # Random
    start_r = random_start(klanten[:], k=4)
    start_r_obj = makespan(start_r, depot)
    t0 = time.time()
    eind_r, eind_r_obj = local_search(start_r, depot, iters=2000)
    tijd_r = round(time.time() - t0, 2)

    resultaten.append([seed, start_g_obj, eind_g_obj, start_r_obj, eind_r_obj, tijd_g, tijd_r])

df = pd.DataFrame(resultaten, columns=["Instance", "Greedy start", "Greedy eind", "Random start", "Random eind", "Tijd G", "Tijd R"])
display(df)

# Simpele vergelijking
plt.plot(df["Instance"], df["Greedy eind"], 'o-', label='Greedy')
plt.plot(df["Instance"], df["Random eind"], 's--', label='Random')
plt.xlabel("Instantie (seed)")
plt.ylabel("Makespan")
plt.title("RQ3: Effect van startoplossing")
plt.legend()
plt.show()
```

```python
# ==============================
# OR5 - Newspaper Delivery
# RQ3: Effect van startoplossing
# met echte dataset (paperboy_instance.xlsx)
# ==============================

import pandas as pd
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# -------------------------------
# Data inlezen
# -------------------------------
df = pd.read_excel("paperboy_instance.xlsx")

print("Data ingelezen:")
print(df.head())

# Depot = eerste rij
depot_x = df.loc[0, "x"]
depot_y = df.loc[0, "y"]
depot = (depot_x, depot_y)

# Klanten = overige rijen
klanten = [(int(x), int(y)) for x, y in zip(df["x"][1:], df["y"][1:])]
print(f"Aantal klanten: {len(klanten)}")

# -------------------------------
# Functies
# -------------------------------

def manhattan(a, b):
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def route_lengte(route, depot):
    dist = 0
    cur = depot
    for klant in route:
        dist += manhattan(cur, klant)
        cur = klant
    return dist

def makespan(routes, depot):
    return max(route_lengte(r, depot) for r in routes)

def greedy_start(klanten, depot, k=4):
    routes = [[] for _ in range(k)]
    klanten_sorted = sorted(klanten, key=lambda c: manhattan(c, depot), reverse=True)
    for klant in klanten_sorted:
        beste_r = None
        beste_obj = 10**9
        for i in range(k):
            temp = [list(r) for r in routes]
            temp[i].append(klant)
            obj = makespan(temp, depot)
            if obj < beste_obj:
                beste_obj = obj
                beste_r = i
        routes[beste_r].append(klant)
    return routes

def random_start(klanten, k=4):
    random.shuffle(klanten)
    routes = [[] for _ in range(k)]
    for i, klant in enumerate(klanten):
        routes[i % k].append(klant)
    return routes

def local_search(routes, depot, iters=2000):
    best = [list(r) for r in routes]
    best_obj = makespan(best, depot)
    k = len(best)
    for _ in range(iters):
        a = random.randrange(k)
        b = random.randrange(k)
        if a != b and len(best[a]) > 0:
            i = random.randrange(len(best[a]))
            klant = best[a][i]
            new_a = best[a][:i] + best[a][i+1:]
            pos = random.randrange(len(best[b])+1)
            new_b = best[b][:pos] + [klant] + best[b][pos:]
            new_routes = [best[x] if x not in (a,b) else (new_a if x==a else new_b) for x in range(k)]
            obj = makespan(new_routes, depot)
            if obj < best_obj:
                best = new_routes
                best_obj = obj
    return best, best_obj

# -------------------------------
# Experiment RQ3
# -------------------------------

random.seed(42)
np.random.seed(42)

# Greedy
start_g = greedy_start(klanten[:], depot, k=4)
start_g_obj = makespan(start_g, depot)
t0 = time.time()
eind_g, eind_g_obj = local_search(start_g, depot, iters=2000)
tijd_g = round(time.time() - t0, 2)

# Random
start_r = random_start(klanten[:], k=4)
start_r_obj = makespan(start_r, depot)
t0 = time.time()
eind_r, eind_r_obj = local_search(start_r, depot, iters=2000)
tijd_r = round(time.time() - t0, 2)

# -------------------------------
# Resultaten
# -------------------------------
print("\n=== RQ3 Resultaten (echte dataset) ===")
print(f"Greedy start  : {start_g_obj}")
print(f"Greedy eind   : {eind_g_obj}  (tijd {tijd_g}s)")
print(f"Random start  : {start_r_obj}")
print(f"Random eind   : {eind_r_obj}  (tijd {tijd_r}s)")

labels = ["Greedy start", "Greedy eind", "Random start", "Random eind"]
waarden = [start_g_obj, eind_g_obj, start_r_obj, eind_r_obj]

plt.bar(labels, waarden, color=["skyblue","dodgerblue","orange","darkorange"])
plt.ylabel("Makespan (tijdseenheden)")
plt.title("RQ3: Effect van startoplossing (echte dataset)")
plt.show()
```
